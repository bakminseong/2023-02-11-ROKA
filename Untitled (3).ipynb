```python
import matplotlib.pyplot as plt
import seaborn as sns

independent_vars = ['medservice01_1', 'medservice01_2', 'service01_1', 'service01_2', 'socservice01_1', 
                    'socservice01_2', 'jobservice01_1', 'jobservice01_2', 'B01014001_1', 'B01014001_2', 
                    'enrolled', 'not_enrolled', 'non_applicable', 'B01003006', 'B01003005_no', 
                    'B01003005_yes','target', 'injurypart', 'injurytype', 'workperiod14', 'disa016']

```

```python
# target별 독립변수 빈도 및 평균 계산
target_distribution_stats = df.groupby("target")[independent_vars].agg(['count', 'mean', 'std'])


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="B01003006", hue="target", kde=True, bins=30)
plt.title("time by Target")
plt.xticks(rotation=45)
plt.legend(title="Target")
plt.show()
```

```python
# 출력 스타일 변경
pd.set_option("display.max_rows", None)  # 모든 행 출력
pd.set_option("display.max_columns", None)  # 모든 열 출력
pd.set_option("display.width", 1000)  # 한 줄에 출력할 최대 너비 증가
pd.set_option("display.colheader_justify", "center")  # 컬럼명 정렬
pd.set_option("display.float_format", lambda x: f"{x:.3f}")  # 소수점 3자리까지 표시


# for문을 사용하여 출력
for var in target_distribution_stats.index:
    print(f"\n--- 독립변수: {var}")
    print(target_distribution_stats.loc[var])
    print("-" * 80)  # 구분선 추가
```

```python
independent_vars = ['medservice01_1', 'medservice01_2', 'service01_1', 'service01_2', 'socservice01_1', 
                    'socservice01_2', 'jobservice01_1', 'jobservice01_2', 'B01014001_1', 'B01014001_2', 
                    'enrolled', 'not_enrolled', 'non_applicable', 'B01003006', 'B01003005_no', 
                    'B01003005_yes','target', 'injurypart', 'injurytype', 'workperiod14']
# 장해등급 뻇음
```

```python


# target과 장해등급별 독립변수 통계 분석
disability_analysis = df.groupby(["target", "disa016"])[independent_vars].agg(['count', 'mean', 'std'])

# 출력 스타일 변경
pd.set_option("display.max_rows", None)  # 모든 행 출력
pd.set_option("display.max_columns", None)  # 모든 열 출력
pd.set_option("display.width", 1000)  # 한 줄에 출력할 최대 너비 증가
pd.set_option("display.colheader_justify", "center")  # 컬럼명 정렬
pd.set_option("display.float_format", lambda x: f"{x:.3f}")  # 소수점 3자리까지 표시

# MultiIndex를 올바르게 처리하여 출력
for tar, disa in disability_analysis.index:
    print(f"\n-- Target: {tar}, 장해등급: {disa}")
    print(disability_analysis.loc[(tar, disa)])
    print("-" * 80)  # 구분선 추가
    

```

```python
for var in independent_vars:
    print(f"\n 변수: {var}")
    print(df.groupby(["target", "disa016"])[var].value_counts(normalize=True))  # 비율 계산 (퍼센트)
    print("-" * 80)
```
